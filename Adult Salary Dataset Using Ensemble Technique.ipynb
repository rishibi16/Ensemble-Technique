```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('adult.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.replace('?', np.nan, inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data['native-country'].value_counts()
```

```python
data.drop('native-country', axis=1, inplace=True)
```

```python
data.relationship.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
data['relationship'] = lb.fit_transform(data['relationship'])
```

```python
data['relationship']
```

```python
data.race.value_counts()
```

```python
data['race'] = lb.fit_transform(data['race'])
```

```python
data.occupation.value_counts()
```

```python
data['occupation'] = data['occupation'].fillna(data['occupation'].mode()[0])
```

```python
data['occupation'] = lb.fit_transform(data['occupation'])
```

```python
data.info()
```

```python
data['workclass'] = data['workclass'].fillna(data['workclass'].mode()[0])
```

```python
data.info()
```

```python
data['workclass'] = lb.fit_transform(data['workclass'])
data['education'] = lb.fit_transform(data['education'])
```

```python
data.head()
```

```python
data['gender'] = data['gender'].apply(lambda x: 1 if x == "Male" else 0)
data['income'] = data['income'].apply(lambda x: 1 if x == "<=50K" else 0)
```

```python
data.head()
```

```python
data['marital-status'] = lb.fit_transform(data['marital-status'])
```

```python
data.head()
```

```python
data.corr()
```

```python
data['fnlwgt'].value_counts()
```

```python
data.drop(['workclass', 'fnlwgt'], axis=1, inplace=True)
```

```python
data.info()
```

```python
X = data.drop('income', axis=1)
X.head()
```

```python
Y = data.income
Y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

## GradientBoosting Classifier

```python
model = GradientBoostingClassifier(learning_rate=0.1, random_state=42).fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

## XGBoost Classifier - An advanced implementation of the gradient boosting algorithm

```python
import xgboost as xgb
```

```python
model1 = xgb.XGBClassifier(learning_rate=0.1, random_state=42).fit(X_train, Y_train)
```

```python
model1.score(X_test, Y_test)
```
